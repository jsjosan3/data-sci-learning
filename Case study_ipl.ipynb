```python
# dataframe of Matches in which teams win by bowling first
bowling_first=ipl[ipl['win_by_wickets']!=0]
bowling_first
```

```python
#x1=list of teams who win by bowling first  (Numerical values)
x1=bowling_first['winner'].value_counts()
x1
```

```python
#y1= name of teams win by bowling first (categorical Values)
y1=bowling_first['winner'].value_counts().keys()
y1
```

```python
 plt.figure(figsize=(7,5))
plt.hist(bowling_first['win_by_wickets'], bins=30, color="purple")
plt.title("matches won by wickets", size= 20, color="purple")
plt.xlabel("Wickets",size= 15, color="purple")
plt.ylabel("Matches win",size= 15, color="purple")
plt.show()
```

```python
sns.displot(x=bowling_first['win_by_wickets'] ,color='purple')
plt.title("matches won by wickets", size= 20, color="purple")
plt.xlabel("Wickets",size= 15, color="purple")
plt.ylabel("Matches win",size= 15, color="purple")
plt.show()
```

```python
# barplot of top 5 teams win by bowling first
plt.figure(figsize=(10,5))
plt.bar(y1[0:5],x1[0:5], color= ["purple", "blue", "red", "pink", "indigo"],  )
plt.title("matches won by wickets", size= 20, )
plt.xlabel("teams",size= 15, )
plt.ylabel("Matches win",size= 15, )
```

```python
plt.figure(figsize=(8,8))
plt.pie(x1[0:10], labels=y1[0:10],autopct=("%0.1f%%"))
plt.title("Winning Percentage of Top 10 Teams by Bowling First", Size= 20, color= "black")
plt.show()
```
