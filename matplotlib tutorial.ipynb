```python
import numpy as np
```

```python
import pandas as pd
```

```python
from matplotlib import pyplot as plt
```

```python
x=np.arange(0,50,5)
x
```

```python
y=x*0.2
y
```

```python
plt.plot(x,y,)
plt.grid(True)
plt.show()
```

```python
x2=y*3
x2
```

```python
plt.plot(x2,y, color="r", linestyle=':', linewidth="4")
plt.title("B/W x & y")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
x=np.arange(5,100,10)
x
```

```python
y1=x*0.5
y1
```

```python
y2=y1*3
y2
```

```python
plt.plot(x,y1, color="r", linestyle=":", linewidth="3")
plt.plot(x,y2, color="g", linestyle="-", linewidth="2")
plt.plot(y1,y2, linestyle="-.", linewidth="3")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.title("Relationship B/W x, Y1, y2")
plt.show()
```

```python
students= {"love": 56, "harjas": 67, "sunny": 65, "preet":45, "Komal": 78, "nav": 68}
students
```

```python
name=list(students.keys())
```

```python
marks=list(students.values())
```

```python
plt.bar(name,marks, color="g" )
plt.title("Marks Comparison")
plt.xlabel("Student Names")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

```python
plt.barh(name,marks, color="r" )
plt.title("Marks Comparison")
plt.xlabel("Student Names")
plt.ylabel("Marks")
#plt.grid(True)
plt.show()
```

```python
x=np.random.randint(1,100,20)
x
```

```python
y=np.random.randint(0,99,20)
y
```

```python
z=np.random.randint(0,99,20)
z
```

```python
plt.scatter(x,y, color="r", marker="+", s= 50)
plt.scatter(x,z, color="g", marker="*", s= 50)
plt.scatter(y,z, color="b", s= 40)
plt.title("random number scattered diagram")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y, color="r", marker="+", s= 50)
plt.title("random number scattered diagram B/W X & Z ")

plt.subplot(1,2,2)
plt.scatter(x,z, color="g", marker="*", s= 50)
plt.title("random number scattered diagram B/W X & Y ")



#plt.subplot(1,3,3)
#plt.scatter(y,z, color="b", s= 40)
plt.title("random number scattered diagram")
plt.show()
```
