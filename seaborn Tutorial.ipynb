# Lineplot In Seaborn

```python
fmri=sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.lineplot(x="timepoint",y="signal", data=fmri, color="g", )
plt.show()
```

```python
sns.lineplot(x="timepoint",y="signal", data=fmri, color="r", hue="event", style="event")
plt.show()
```

we can add color, hue, style, markers in Lineplot as mentioned below

```python
sns.lineplot(x="timepoint",y="signal", data=fmri, color="r", hue="region", style="region", markers=True)
plt.show()
```

# Barplot in Seaborn

```python
import pandas as pd
```

```python
tips=sns.load_dataset('tips')
```

```python
tips
```

```python
sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(5,5))
sns.barplot(y="tip", x="sex", data=tips)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="sex", y="total_bill", data= tips, palette= "vlag")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="day", y="total_bill", data= tips)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="day", y="tip", data= tips)
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="sex", y="total_bill", data= tips, hue="day")
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.barplot(x="time", y="total_bill", data=tips, hue="day")
plt.show()
```

# Distplot/Histogram  in Seaborn

'anagrams', 'anscombe', 'attention', 'brain_networks', 'car_crashes', 'diamonds', 'dots', 'exercise', 'flights', 'fmri', 'gammas', 'geyser', 'iris', 'mpg', 'penguins', 'planets', 'tips', 'titanic'

```python
diamonds=sns.load_dataset('diamonds')
diamonds
```

```python
sns.distplot(diamonds["price"])
plt.show()
```

```python
sns.distplot(diamonds["price"], hist=False, color= "r")
plt.show()
```

```python
sns.histplot(diamonds["price"], bins=20, color= "g")
plt.show()
```

# jointplot in Seaborn   (scater+hist)

```python
iris
```

```python
sns.jointplot(x='sepal_length', y="petal_length", data=iris, hue="species")
plt.show()
```

```python
sns.jointplot(x='sepal_width', y="petal_width", data=iris, hue="species")
plt.show()
```

```python
sns.jointplot(x='sepal_width', y="petal_width", data=iris,  kind="reg")
plt.show()
```

```python
sns.jointplot(x='sepal_length', y="petal_length", data=iris, color="purple" ,kind="reg",) #hue="species")
plt.show()
```
